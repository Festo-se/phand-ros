# Loomia sensor evaluation - Round 2

Evaluation of Loomia sensors from prototype round 2. Round 1 sensor shad very low sensitivity. Round 2 seem way more sensitive, this is an attempt to quantify their performance.

```python
# Library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data from .xls file
DATA_FOLDER = 'data'
DATA_SAMPLE = '/sens_eval_2A.xlsx'
```

## Data from sensor unworn, no pressure applied

```python
data_sample = pd.read_excel(DATA_FOLDER+DATA_SAMPLE, sheet_name='1-unworn')
data_sample = data_sample[['A1','A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3']]
data_sample.head(5)
```

```python
print(data_sample.describe())
```

```python
# Time plot
fig, ax = plt.subplots(figsize=(16,4))
ax.plot(data_sample)
colourmap = plt.cm.gist_ncar
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 0.9, 9))))

ax.set(xlabel='time steps', ylabel='signal (analog counts)',
       title='data sample: 3-worn robot')
ax.grid()
labels = ['A1','A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3']
ax.legend(labels)

#fig.savefig("test.png")
plt.show()
```

```python
plt.figure(figsize=(16,4))
data_sample.describe().boxplot()
```
