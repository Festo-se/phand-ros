## Zone overview

```python
# Load data from .csv file
DATA_FOLDER = '../data/works-like/'
data_files = sorted([f for f in os.listdir(DATA_FOLDER) if f.endswith('.csv')]) 

fig, axs = plt.subplots(3, 3, figsize=(16,16))

for n, file in enumerate(sorted(data_files)):
    
    df = pd.read_csv(DATA_FOLDER+file, header=None)
    df = df.drop([0]).drop(columns=[0]).astype(np.float16).astype(int)
    map_matrix2(fig, axs, df, file, n, what='max', remap='yes')
```

## All combined data

```python
# Load data from .csv file
DATA_FOLDER = '../data/works-like/'
data_files = sorted([f for f in os.listdir(DATA_FOLDER) if f.endswith('.csv')]) 

df_all = pd.DataFrame()

for n, file in enumerate(sorted(data_files)):
    
    df = pd.read_csv(DATA_FOLDER+file, header=None)
    df = df.drop([0]).drop(columns=[0]).astype(np.float16).astype(int)
    
    if file != 'all.csv':
        df_all = pd.concat([df_all,df])
        print(file+' added to all.csv')
        
df_all.to_csv(DATA_FOLDER+'all.csv', encoding='utf-8')
map_matrix(df_all, sample_name, what='max', remap='yes')
```
